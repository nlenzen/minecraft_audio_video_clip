```python
# Wrapper classes for models
# All wrapper classes have to implement a get_embeddings method

class ASTWrapper(ASTModel):
  def add_sr(self, sr):
    self.sr = sr

  def add_name(self, name):
    self.name = name

  def get_embeddings(self, data_wrapper, device):
    input_tdim = 1024
    embeddings = []
    for _, sample, _ in tqdm.tqdm(data_wrapper):
      features = self.make_features(sample.unsqueeze(0))
      features = features.expand(1, input_tdim, 128)           # reshape the feature
      features = features.to(device)
      embedding = self.forward(features).squeeze().detach()
      embedding = embedding.cpu().numpy()
      embeddings.append(embedding)

    return embeddings

  def make_features(self, sample, mel_bins=128, target_length=1024):
    fbank = torchaudio.compliance.kaldi.fbank(
        sample, htk_compat=True, sample_frequency=self.sr, use_energy=False,
        window_type='hanning', num_mel_bins=mel_bins, dither=0.0, frame_shift=10)

    n_frames = fbank.shape[0]

    p = target_length - n_frames
    if p > 0:
        m = torch.nn.ZeroPad2d((0, 0, 0, p))
        fbank = m(fbank)
    elif p < 0:
        fbank = fbank[0:target_length, :]

    fbank = (fbank - (-4.2677393)) / (4.5689974 * 2)
    return fbank


class Whisper_atWrapper:
  model = None

  def __init__(self, name, sr):
    self.name = name
    self.sr = sr

    self.N_FFT = 400
    self.HOP_LENGTH = 160
    self.CHUNK_LENGTH = 30
    self.n_samples = self.CHUNK_LENGTH * sr
    self.N_FRAMES = self.exact_div(self.n_samples, self.HOP_LENGTH)

  def get_embeddings(self, data_wrapper, device):
    embeddings = []
    for _, sample, _ in tqdm.tqdm(data_wrapper):
      features = self.log_mel_spectrogram(sample.unsqueeze(0), device, padding=self.n_samples)
      features = self.pad_or_trim(features)
      audio_features, audio_features_for_at = self.model.encoder(features)
      embedding = self.model.at_model(audio_features_for_at)
      embedding = embedding[0]
      #embedding = embedding.mean(dim=0).detach()
      embedding = embedding.cpu().numpy()
      embeddings.append(embedding)

    return embeddings

  def mel_filters(sshow cluster boundarieself, device, n_mels=80):
    assert n_mels == 80, f"Unsupported n_mels: {n_mels}"
    path = '/content/drive/MyDrive/Bachelorarbeit/source/mel_filters.npz'
    with np.load(path) as f:
        return torch.from_numpy(f[f"mel_{n_mels}"]).to(device)

  def log_mel_spectrogram(self, sample, device, padding=0, n_mels=80):
    if device is not None:
        sample = sample.to(device)
    if padding > 0:
        sample = F.pad(sample, (0, padding))
    window = torch.hann_window(self.N_FFT).to(sample.device)
    stft = torch.stft(sample, self.N_FFT, self.HOP_LENGTH, window=window, return_complex=True)
    magnitudes = stft[..., :-1].abs() ** 2

    filters = self.mel_filters(device, n_mels)
    mel_spec = filters @ magnitudes

    log_spec = torch.clamp(mel_spec, min=1e-10).log10()
    log_spec = torch.maximum(log_spec, log_spec.max() - 8.0)
    log_spec = (log_spec + 4.0) / 4.0
    return log_spec

  def pad_or_trim(self, array, *, axis: int = -1):
    length = self.N_FRAMES
    if torch.is_tensor(array):
      if array.shape[axis] > length:
        array = array.index_select(
            dim=axis, index=torch.arange(length, device=array.device)
        )

      if array.shape[axis] < length:
        pad_widths = [(0, 0)] * array.ndim
        pad_widths[axis] = (0, length - array.shape[axis])
        array = F.pad(array, [pad for sizes in pad_widths[::-1] for pad in sizes])
    else:
      if array.shape[axis] > length:
        array = array.take(indices=range(length), axis=axis)

      if array.shape[axis] < length:
        pad_widths = [(0, 0)] * array.ndim
        pad_widths[axis] = (0, length - array.shape[axis])
        array = np.pad(array, pad_widths)

    return array

  def exact_div(self, x, y):
    assert x % y == 0
    return x // y

```

```python
# Helper functions

def plot_data(data, labels, clf, img_path, method):
  class_to_num = {label: i for i, label in enumerate(set(labels))}
  numeric_labels = [class_to_num[label] for label in labels]

  unique_labels = np.unique(labels)
  colors = plt.cm.turbo(np.linspace(0, 1, len(unique_labels)))
  unique_labels = unique_labels[::-1]
  colors = colors[::-1]

  markers = ['x', '^', 'v', '<', '>', 's', 'p', '*', 'h', 'X', '.', 'd']

  plt.clf()
  plt.figure(figsize=(10, 8))

  centers = clf.cluster_centers_
  c_length = len(centers)

  centers = np.append(centers, [[999, 999], [-999, 999], [999, -999], [-999, -999]], axis=0)
  center_colors = plt.cm.rainbow(np.linspace(0, 1, len(centers) + 1))
  vor = Voronoi(centers)
  fig = voronoi_plot_2d(vor, plt.gca(), show_vertices=False, s=1, show_points=False, line_width=1, line_alpha=0.15)

  for i, region in enumerate(vor.regions):
    if not -1 in region:
        polygon = [vor.vertices[i] for i in region]
        plt.fill(*zip(*polygon), color=center_colors[i], alpha=0.1)

  for i, label_value in enumerate(unique_labels):
    indices = np.where(np.array(labels) == label_value)
    marker = markers[i % len(markers)]
    plt.scatter(data[indices, 0], data[indices, 1], label=f'{label_value}', color=colors[i], marker=marker, s=10, alpha=1)

  centers = centers[:c_length]
  plt.scatter(centers[:,0], centers[:,1], marker='o', color='black', s=12)
  for i, center in enumerate(centers):
    plt.annotate('Cluster {}'.format(i), (center[0], center[1]), fontsize=12)

  plt.xlim(data[:, 0].min() - 1, data[:, 0].max() + 1)
  plt.ylim(data[:, 1].min() - 1, data[:, 1].max() + 1)

  plt.title(f'Clustering of Encoded Audio Representations {method=}')
  plt.legend(prop={"size": 10})
  plt.savefig(img_path)
  plt.show()
  # Close the plot to release resources
  #plt.close()


def plot_deviation(data1, data2, savepath):
  xx = [data1[:, 0], data2[:,0]]
  yy = [data1[:, 1], data2[:,1]]

  plt.clf()
  plt.figure(figsize=(10, 8))

  plt.plot(xx, yy, 'k-', linewidth=1)

  plt.scatter(data1[:,0], data1[:,1], color='b', marker='o', label='Dataset 1')
  plt.scatter(data2[:,0], data2[:,1], color='r', marker='o', label='Dataset 2')
  plt.title('Deviation of samples with different length')

  plt.savefig(savepath)
  plt.plot()


def plot_bars(data1, data2, savepath):
  fig, axs = plt.subplots(len(data1))
  fig.suptitle('Comparison of embedding dimensions')
  x = np.arange(len(data1[0]))

  for i, ax in enumerate(axs):
    ax.set_xlabel('Dimensions')

    ax.bar(x - 0.05, data1[i], color='r', width=0.1)
    ax.bar(x + 0.05, data2[i], color='b', width=0.1)

  plt.savefig(savepath)
  plt.plot()
```

```python
# Test Pipeline

def pipeline(data_wrapper, models, dirname, device):
#  device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
  for model_wrapper in models:
    print('Processing model ' + model_wrapper.name)
    model_data_wrapper = data_wrapper.resample(model_wrapper.sr)
    with torch.no_grad():
      embeddings = model_wrapper.get_embeddings(model_data_wrapper, device)

    # calculate t-SNE
    embeddings = np.array(embeddings)

    tsne = TSNE(n_components=2, random_state=42)
    embedded_tsne = tsne.fit_transform(embeddings)

    # calculate UMAP
    umap = UMAP(n_components=2, random_state=42)
    embedded_umap = umap.fit_transform(embeddings)

    # clustering for tsne
    #n_clusters = len(set(model_data_wrapper.labels))
    n_clusters = 20

    kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init="auto").fit(embedded_tsne)
    clustering_tsne = kmeans.labels_

    # add cluster name to sample names
    for i, name in enumerate(model_data_wrapper.names):
      model_data_wrapper.names[i] = ''.join((str(clustering_tsne[i]), "/", name))

    # plotting setup aValueError: Buffer dtype mismatch, expected 'double' but got 'float'nd plotting tsne
    base_path = '/content/drive/MyDrive/Bachelorarbeit/results'
    base_path = os.path.join(base_path, dirname)
    img_path = os.path.join(base_path, 'images', model_wrapper.name) + '_tsne.png'
    os.makedirs(os.path.dirname(img_path), exist_ok=True)
    plot_data(embedded_tsne, model_data_wrapper.labels, kmeans, img_path, 'tsne')

    # save data based on the tsne clusters
    save_path = os.path.join(base_path, model_wrapper.name, 'clusters', 'tsne')
    util.save_audio(save_path, model_data_wrapper)

    # clustering for umap
    kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init="auto").fit(embedded_umap)
    clustering_umap = kmeans.labels_

    # add new cluster name to sample names
    for i, name in enumerate(model_data_wrapper.names):
      base_name = name.split('/', 1)[1]
      model_data_wrapper.names[i] = ''.join((str(clustering_umap[i]), "/", base_name))

    # Plotting umap
    img_path = os.path.join(base_path, 'images', model_wrapper.name) + '_umap.png'
    plot_data(embedded_umap, model_data_wrapper.labels, kmeans, img_path, 'umap')

    # save data based on the umap clusters
    save_path = os.path.join(base_path, model_wrapper.name, 'clusters', 'umap')
    util.save_audio(save_path, model_data_wrapper)
```
